## Distribución Uniforme
* Se categoriza por generar números aleatorios equiprobables (misma probabilidad para todos)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
a = 1
b = 100
n = 20000000
data = np.random.uniform(a, b, n)
```

```python
# Pintamos un Histograma de los número generados
# Cuantos más generemos más se asemejará a una Distribución Uniforme
# Misma probabilidad de caer en un intervalo u otro
plt.hist(data)
```

## Distribución Normal
* Campana de Gauss
* Parámteros que definen la distribución normal:
    * <b>Media: </b> es igual a 0. Es decir, los valores están centrados entorno al cero
        * Tenemos un <b>50% de probabilidades</b> de caer en un lado u otro de la campana
    * <b>Desviación Típica/Estándar: </b> es igual a 1. Que es dónde la curva deja de estar "triste" a estar "contenta"
        * Entre <b>-1 y 1 de las desviación típica </b> se encuentran el <b>70%</b> de los datos
        * Lo "normal" es estar entorno al promedio

```python
# Generamos 100 valores con distribución normal
n = 5000000
data2 = np.random.randn(n)
```

```python
# Pintamos los valores generados
# El eje x tenemos el rango de nuestros datos
# Y la altura corresponde con nuestros datos
x = range(n)
plt.plot(x, data2)
```

```python
plt.hist(data2)
```

```python
# Función de distribución acumulada
plt.plot(x, sorted(data2))
```

## La simulación de Monte Carlo


Generamos dos números aleatorios (x,y) entre 0 y 1 'n' veces <br>
Calcularemos <b> X ^ 2 + Y ^ 2 </b>
* Si el valor es inferior a 1 --> estamos dentro del círculo
* Si el valor es superior a 1 --> estamos fuera del círculo

Calculamos el número total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener un aproximación de la probabilidad de caer dentro del círculo. <br>
Usamos dicha probabilidad para aproximar el valor de π. <br>
Repetimos el experimento un número suficiente de veces (por ejemplo 1000 -> 'n_exp'), para obtener diferentes aproximaciones de π. <br>
Calculamos el promedio de los 'n_exp' experimentos anteriores para dar un valor final de π.

```python
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []

    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0, 1, n).tolist()
        y = np.random.uniform(0, 1, n).tolist()

        for j in range(n):
            z = np.sqrt(x[j] ** 2 + y[j] ** 2)

            if z <= 1:
                value += 1

        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value


    pi = pi_avg / n_exp
    
    print(pi)

    fig = plt.plot(pi_value_list)
    
    return (pi, fig)
```

```python
pi_montecarlo(20000, 400)
```
