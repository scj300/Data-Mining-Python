# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Dos categrías: Género y Nivel de Ingresos
gender = ['Male', 'Female']
income = ['Poor', 'Middle Class',  'Rich']
```

```python
n = 500

gender_data = []
income_data = []

# 'n' datos aleatorios de ambas categrorías
for i in range(n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:7]
```

```python
income_data[:7]
```

```python
height = 160 + 30 * np.random.randn(n)       # Media de las alturas: 160cm con una dispersión de 30
weight = 65 + 25 * np.random.randn(n)        # Media de los pesos: 65kg con una dispersión de 25
age = 30 + 12 * np.round(np.random.randn(n)) # Media de los años: 30 con una dispersión de 12
income = 18000 + 3500 * np.random.randn(n)   # Media de ingresos: 18000 € con una dispersión de 3500
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.head()
```

## Agrupación de Datos

```python
# No rompe los datos originales
# Crea un objeto con dos atributos: el nombre y el grupo
grouped_gender = data.groupby('Gender')
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
# Obtener solo el grupo de mujeres
grouped_gender.get_group('Female')[:7]
```

```python
# Por Género y Estado Socioeconómico
# Creará 6 grupos: 2 Por las categorías de Género y 3 por las catgerorías status económico = 2 * 3 = 6
double_group = data.groupby(['Gender', 'Economic Status']) 
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
# Suma de todos los atributos de nuestro data-set
double_group.sum()
```

```python
# Promedio
double_group.mean()
```

```python
# Función size --> tamaño
double_group.size()
```

```python
# Describe()
# Los estadísticos para nuestras categorías y variables
double_group.describe()
```

```python
# Obtenemos una variable y le aplicamos la función de agregación deseada
grouped_income = double_group['Income']
grouped_income.describe()
```

```python
# Agregamos las funciones que nos interesen
double_group.aggregate(
    {
        'Income' : np.sum,
        'Age' : np.mean,
        'Height' : np.std
    }
)
```

```python
# Media de la altura / la desviación típica
double_group.aggregate(
    {
        'Age' : np.mean,
        'Height' : lambda h: np.mean(h) / np.std(h)
    }
)
```

```python
# Aplicar [np.sum, np.mean, np.std] a todas las variables
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Filtrado de Datos

```python
# Suma de las edades por grupos que superan los 2400 años
double_group['Age'].filter(lambda x: x.sum() > 2400)[:7]
```

## Transformación de Variables

```python
# X simboliza una de las columnas
zscore = lambda x : (x - x.mean()) / x.std() 
```

```python
z_group = double_group.transform(zscore)
```

```python
plt.hist(z_group['Age'])
```
